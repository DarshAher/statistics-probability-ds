## üîç What Are Probability Distributions?
A **probability distribution** tells us how likely outcomes are distributed over a range. It's how we model randomness ‚Äî like how many people show up in an hour, or how long until the next customer arrives.

## üî¢ Discrete vs. Continuous Distributions
- **Discrete**: Only specific values (e.g., number of orders = 5)
- **Continuous**: Any value within a range (e.g., time spent = 7.2 mins)

At Cafe Zenith:
- Discrete ‚Üí number of customers (Poisson), success/failure orders (Binomial)
- Continuous ‚Üí wait times (Exponential), coffee ratings (Normal)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
sns.set(style='whitegrid')
np.random.seed(42)
```

## üìà Normal Distribution: Modeling Coffee Ratings
The normal distribution (bell curve) is ideal for modeling continuous symmetric data like customer satisfaction ratings.

```python
# Generate normal data
ratings = np.random.normal(loc=4.0, scale=0.3, size=1000)

# Plot PDF
sns.histplot(ratings, kde=True, stat="density", bins=30, color='skyblue')
x = np.linspace(3, 5, 100)
plt.plot(x, stats.norm.pdf(x, loc=4.0, scale=0.3), 'r', lw=2, label='PDF')
plt.title("Normal Distribution - Customer Ratings")
plt.xlabel("Rating (1-5)")
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm
sm.qqplot(ratings, line='s')
plt.title("Q-Q Plot - Are Ratings Normally Distributed?")
plt.show()
```

## üßÆ Binomial Distribution: Espresso Orders per Customer
What‚Äôs the chance a customer orders an espresso (success) out of 5 visits?
Let‚Äôs assume probability of ordering = 0.3.

```python
# Simulate Binomial Trials
binom_data = np.random.binomial(n=5, p=0.3, size=1000)
sns.histplot(binom_data, stat="probability", discrete=True, bins=range(7), color='lightgreen')
plt.title("Binomial Distribution - Espresso Orders per 5 Visits")
plt.xlabel("Espresso Orders")
plt.show()
```

## ‚è∞ Poisson Distribution: Customer Arrivals per Hour
Poisson models count of rare events. Let‚Äôs say Cafe Zenith sees ~4 customers per hour.

```python
poisson_data = np.random.poisson(lam=4, size=1000)
sns.histplot(poisson_data, stat="probability", discrete=True, color='orange')
plt.title("Poisson Distribution - Customer Arrivals Per Hour")
plt.xlabel("Arrivals")
plt.show()
```

## üé≤ Uniform Distribution: Random Table Assignment
Let‚Äôs say you assign customers to tables 1 through 10 at random. Each table has equal likelihood.

```python
uniform_data = np.random.randint(1, 11, size=1000)
sns.histplot(uniform_data, bins=10, discrete=True, color='purple')
plt.title("Uniform Distribution - Table Assignments")
plt.xlabel("Table Number")
plt.show()
```

## ‚è≥ Exponential Distribution: Time Between Orders
Exponential distributions model waiting time between independent events ‚Äî like new customer arrivals.

```python
exp_data = np.random.exponential(scale=3, size=1000)
sns.histplot(exp_data, bins=30, kde=True, color='coral')
plt.title("Exponential Distribution - Time Between Orders (in mins)")
plt.xlabel("Minutes")
plt.show()
```
